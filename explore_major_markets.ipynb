### 1. Does the average price of a metro area have a correlation with its features?

<br>
$H_0$: There is no correlation between total listings and the average price

$H_a$: There is a correlation between total listings and the average price

```python
α = 0.05

listing = monthly_sa['total_listings']
home_value = monthly_sa['average_price']

corr, p_value = pearsonr(listing, home_value)

if p_value < α:
    print("Reject the Null Hypothesis. There is a correlation between total listings and average price.")
else:
    print("Fail to reject the Null Hypothesis. There is no correlation between total listings and average price.")
```

```python
corr,p_value
```

```python
α = 0.05

columns = ['sales','dollar_volume', 'total_listings','months_inventory']
correlation_results = []

for col in columns:
    listing = monthly_sa[col]
    home_value = monthly_sa['average_price']

    corr, p_value = pearsonr(listing, home_value)
    
    correlation_results.append({'feature': col, 'correlation': corr, 'p-value': p_value})

dff = pd.DataFrame(correlation_results)
dff['significance']=dff['p-value'] < α
dff.sort_values('correlation', ascending = False, ignore_index = True)
```

### 2. Do metro areas have a relationship with other metro areas?


```python
α = 0.05

metros = ['austin','houston','dallas']
correlation_results = []

for metro in metros:
    listing = df[metro]
    home_value = monthly_sa['average_price']

    corr, p_value = pearsonr(listing, home_value)
    
    correlation_results.append({'feature': metro, 'correlation': corr, 'p-value': p_value})

dff = pd.DataFrame(correlation_results)
dff['significance']=dff['p-value'] < α
dff.sort_values('correlation', ascending = False, ignore_index = True, inplace=True)
```

```python
def comparing_markets(mkt1,mkt2,mkt3,mkt4, keys = ['name1', 'name2', 'name3', 'name4']):
    markets = [mkt1,mkt2,mkt3,mkt4]
    
    concat_markets = []
    for market in markets:
        df = time_df(market)
        concat_markets.append(df['average_price'])  

    df = pd.concat(concat_markets, axis=1, 
    keys = keys)
    
    return df
```

```python
test = comparing_markets(mon_aus,mon_dal,mon_hou,monthly_sa,
                         keys = ['austin', 'dallas', 'houston', 'san antonio'])
```

```python
def create_market_subplots(markets, titles):
    num_markets = len(markets)
    num_rows = (num_markets - 1) // 2 + 1
    num_cols = min(num_markets, 2)
    
    fig, axs = plt.subplots(num_rows, num_cols, figsize=(10, 8))

    for i, market in enumerate(markets):
        row = i // num_cols
        col = i % num_cols
        ax = axs[row, col]
        sns.regplot(data=market, x='sales', y='average_price', line_kws={'color': 'red'}, ax=ax)
        ax.set_title(titles[i])

    plt.tight_layout()
    plt.show()
```

```python
markets = [mon_dal, mon_aus, mon_hou, monthly_sa]
titles = ['Monthly Dallas', 'Monthly Austin', 'Monthly Houston', 'Monthly San Antonio']

create_market_subplots(markets, titles)
```

```python
def plot_market_home_prices(df, title, x_label, y_label, figsize=(12, 8), palette=None, linestyle='-', dashes=False):
    sns.set_style('whitegrid')
    sns.set_palette(palette)

    plt.figure(figsize=figsize)
    sns.lineplot(data=df, linestyle=linestyle, dashes=dashes)

    plt.title(title, fontsize=20)
    plt.xlabel('')
    plt.ylabel(f'{y_label} ($)', fontsize=17)  # Include currency symbol in y-axis label
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)

    plt.legend(df.columns, fontsize=15)

    sns.despine()
    plt.tight_layout()
    
    plt.show()
```
