```python
sns.lmplot(data = monthly_sa, x = 'sales', y = 'average_price', line_kws= {'color': 'red'})
```

```python
plt.subplot(2, 2, 1)
sns.regplot(data=mon_dal, x='sales', y='average_price', line_kws={'color': 'red'})
plt.title('Monthly Dallas')

plt.subplot(2, 2, 2)
sns.regplot(data=mon_aus, x='sales', y='average_price', line_kws={'color': 'red'})
plt.title('Monthly Austin')

plt.subplot(2, 2, 3)
sns.regplot(data=mon_hou, x='sales', y='average_price', line_kws={'color': 'red'})
plt.title('Monthly Houston')

plt.subplot(2, 2, 4)
sns.regplot(data=monthly_sa, x='sales', y='average_price', line_kws={'color': 'red'})
plt.title('Monthly San Antonio')

plt.tight_layout()
plt.show()
```
